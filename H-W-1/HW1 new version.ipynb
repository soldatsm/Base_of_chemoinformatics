```python
import pubchempy as pcp
import pandas as pd
```

```python
cid = pd.read_table("/home/aster/Base_of_chemoinformatics/H-W-1/chem_id_pubmed.txt", names=["CID"])
```

```python
frame = pd.DataFrame({"CID":cid["CID"],
    "SMILES":[pcp.Compound.from_cid(x).canonical_smiles for x in cid["CID"]],
    "MolecularFormula":[pcp.Compound.from_cid(i).molecular_formula for i in cid["CID"]],
    "FingerPrint2D":[pcp.Compound.from_cid(j).fingerprint for j in cid["CID"]]})
```

MACCkey, MorganFingerpeint, RDfingerprint



```python
from rdkit.Chem import MACCSkeys
from rdkit import DataStructs
```

```python
#MACCSkey
MACCS_fp = [MACCSkeys.GenMACCSKeys(x) for x in mol]
```

```python
compare_MACCSkey = []
for i in range(0, len(MACCS_fp)):
    for j in range(i+1, len(MACCS_fp)):
        score = DataStructs.FingerprintSimilarity(MACCS_fp[i], MACCS_fp[j])
        compare_MACCSkey.append(round(score, 4))
        print("Similarity between", 
             frame.CID[i], "and",
             frame.CID[j], "=",round(score, 4) )
```

```python
from rdkit.Chem import AllChem
```

```python
Morgan_fp = [AllChem.GetMorganFingerprintAsBitVect(x, 2) for x in mol]
```

```python
compare_Morgan_fp =[]
for i in range(0, len(Morgan_fp)):
    for j in range(i+1, len(Morgan_fp)):
        score2 = DataStructs.FingerprintSimilarity(Morgan_fp[i], Morgan_fp[j])
        compare_Morgan_fp.append(round(score2, 4))
        print("Similarity between", 
             frame.CID[i], "and",
             frame.CID[j], "=", round(score2, 4))
```

```python
RDKit_fp = [Chem.RDKFingerprint(x) for x in mol]
```

```python
compare_RDkit_fp = []
for i in range(0, len(RDKit_fp)):
    for j in range(i+1, len(RDKit_fp)):
        score3 = DataStructs.FingerprintSimilarity(RDKit_fp[i], RDKit_fp[j])
        compare_RDkit_fp.append(round(score3, 4))
        print("Similarity between", 
             frame.CID[i], "and",
             frame.CID[j], "=", round(score3, 4))
```

```python
from matplotlib import pyplot as plt
```

```python
plt.hist(compare_Morgan_fp, alpha = 0.33, label="Morgan")
plt.hist(compare_MACCSkey, alpha = 0.33, label="MACCS")
plt.hist(compare_RDkit_fp, alpha = 0.33, label="RDkit fp")
plt.ylabel("Number of Compares")
plt.xlabel("Values")
plt.legend(loc='upper right')
```
